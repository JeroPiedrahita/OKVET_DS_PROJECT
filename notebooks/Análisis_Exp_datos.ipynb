**An치lisis exploratorios de datos**

```python
import os
import pandas as pd
```

Funcion de analisis

```python
def analizar_hoja(df, nombre):
    """
    Muestra un an치lisis b치sico de un DataFrame de pandas.

    Args:
        df (pd.DataFrame): El DataFrame a analizar.
        nombre (str): Nombre descriptivo del DataFrame.
    """
    print(f"\n游늵 An치lisis de: {nombre}")
    print(f"Forma: {df.shape}")
    print(f"Columnas: {df.columns.tolist()}\n")
    
    print("Tipos de datos:")
    print(df.dtypes, "\n")
    
    print("Valores nulos por columna:")
    print(df.isnull().sum(), "\n")
    
    print("Valores 칰nicos por columna:")
    print(df.nunique(), "\n")
    
    print("Ejemplo de datos:")
    try:
        from IPython.display import display
        display(df.head())
    except ImportError:
        print(df.head())
```

Extraccion de Hojas de Excel

```python
carpeta = "/home/jeropf/OKVET/OKVET_DS_PROJECT/data"  # Cambia esto por la ruta deseada

for archivo in os.listdir(carpeta):
    if archivo.lower().endswith('.xlsx'):
        ruta_archivo = os.path.join(carpeta, archivo)
        try:
            excel = pd.ExcelFile(ruta_archivo)
            for hoja in excel.sheet_names:
                try:
                    df = pd.read_excel(ruta_archivo, sheet_name=hoja)
                    analizar_hoja(df, f"{archivo} - {hoja}")
                except Exception as e:
                    print(f"仇 Error leyendo hoja '{hoja}' en '{archivo}': {e}")
        except Exception as e:
            print(f"仇 Error abriendo archivo '{archivo}': {e}")
```
